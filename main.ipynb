```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
print("SEPARATE ARRAYS:\n",data['INDICATOR'].unique(), data['Indicator'].unique())

indicator_dict={}
for IND,ind in zip(data['INDICATOR'].unique(), data['Indicator'].unique()):
    indicator_dict[IND]=ind
    
print("\n\nCOMBINED DICTIONARY\n",indicator_dict)
def ind_info(INDs):
    ops=[]
    for IND in INDs:
        ops.append(indicator_dict[IND])
    return ops
```

```python
gdp_target=pd.read_csv("gdp_per_capita.csv", delimiter='\t')
gdp_target_1=pd.DataFrame([gdp_target['Country'],gdp_target['2015']]).T
gdp_target_1
```

```python
countries_parent=gdp_target_1['Country'].values
gdp_vals_parent=gdp_target_1['2015'].values
```

```python
countries=data['Country'].unique()
```

```python
countries
```

```python
cont_gdp_dict_parent={}
for c,g in zip(countries_parent, gdp_vals_parent):
    cont_gdp_dict_parent[c]=g
```

```python
cont_gdp_dict_parent
```

```python
gdp_vals_string=[]
for con in countries[:-1]:
    gdp_vals_string.append(cont_gdp_dict_parent[con])
gdp_vals=[]
for val in gdp_vals_string:
    gdp_vals.append(float(val.replace(',','')))
gdp_vals=np.array(gdp_vals)

gdp_vals
```

```python
len(gdp_vals)
```

```python
drop_these=['LOCATION','Indicator','Measure','Inequality','MEASURE','Reference Period Code','Reference Period','Flags','Flag Codes',
                  'PowerCode Code','PowerCode','Unit','Unit Code','Country','INDICATOR','INEQUALITY']
keep_these=['Value']
data_country_index=pd.DataFrame(data.drop(columns=drop_these).values,index=data['Country'].values, columns=keep_these)
data_country_index.head()
```
